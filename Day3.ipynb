* 연도를 입력했을때, 해당 연도의 띠를 구하는 함수(def)를 생성하시오.

```python
year = int(input("연도를 입력하시오: "))
```

```python
def year_function(data):
    if(data%12==0):
        print("신(원숭이)띠")
    elif(data%12==1):
        print("유(닭)띠")
    elif(data%12==2):
        print("술(개)띠")
    elif(data%12==3):
        print("해(돼지)띠")
    elif(data%12==4):
        print("자(쥐)띠")
    elif(data%12==5):
        print("축(소)띠")
    elif(data%12==6):
        print("인(호랑이)띠")
    elif(data%12==7):
        print("묘(토끼)띠")
    elif(data%12==8):
        print("진(용)띠")
    elif(data%12==9):
        print("사(뱀)띠")
    elif(data%12==10):
        print("오(말)띠")
    elif(data%12==11):
        print("미(양)띠") 
```

```python
year_function(2019)
```

```python
def ddi(data):
    result=data%12
    dic1={0:"원숭이", 1:"닭", 2:"개", 3:"돼지", 4:"쥐", 5:"쥐", 6:"호랭이", 7:"토끼",8:"용",9:"뱀",10:"말",11:"양"}
    return dic1[result]
```

```python
def div_date(row):
    return row[0:2]

df1["계약월"]=df1["계약일"].apply(div_date)
```

```python
df1["계약일(clean)"]=pd.to_datetime(df1["계약일"])
```

```python
df1["계약연도"]=df1["계약일(clean)"].dt.year # .dt는 날짜의 함수
```

```python
df1["계약월"]=df1["계약일(clean)"].dt.month
df1["계약주차"]=df1["계약일(clean)"].dt.week
df1["계약요일"]=df1["계약일(clean)"].dt.weekday_name # 계약 요일
```
